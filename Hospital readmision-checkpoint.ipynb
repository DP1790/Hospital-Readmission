### Removing outliers using zscore

```python
#Function to choose the right threshold
def threshold(z,d):
    for i in np.arange(3,5,0.2):
        data=d.copy()
        data=data[(z<i).all(axis=1)]
        loss=(d.shape[0]-data.shape[0])/d.shape[0]*100
        print('With threshold {} data loss is {}%'.format(np.round(i,1),np.round(loss,2))) 
```

```python
from scipy.stats import zscore
z = np.abs(zscore(df1))
threshold(z,df1)
```

- As we have seen above few of the columns are categorical & few columns are having outliers hwoever if we apply zscore we would almost loss 100% of the data. So lets keep it & build the model.

```python
## We have highly skewed few of the columns, Let's try to remove them.
df1['NT-proBNP'] = np.sqrt(df1['NT-proBNP'])
df1['LengthOfStay'] = np.sqrt(df1['LengthOfStay'])
df1['RespiratoryRateMean'] = np.sqrt(df1['RespiratoryRateMean'])
df1['RespiratoryRateMax'] = np.sqrt(df1['RespiratoryRateMax'])
df1['CardiacTroponin'] = np.sqrt(df1['CardiacTroponin'])
df1['BNP'] = np.sqrt(df1['BNP'])
```

### Finding best Random_ state

```python
maxAccu = 0
maxRS = 0
for i in range(1,150):
    x_train, x_test, y_train, y_test = train_test_split(X_scaled,y_new, test_size = .24, random_state= i)
    rf= RandomForestClassifier()
    rf.fit(x_train,y_train)
    predrf = rf.predict(x_test)
    acc = accuracy_score(y_test, predrf)
    if acc > maxAccu:
        maxAccu = acc
        maxRS = i
        
print("Best accuracy is ", maxAccu, "on Random_state", maxRS)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y_new ,test_size= .23, random_state=maxRS)
```
